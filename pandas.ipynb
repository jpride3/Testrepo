```python
import pandas as pd
data = {
    "name": ["jack", "john", "james", "ann"],
    "age": [23, 25, 40, 35],
    "counties": ["kajiado", "nyeri", "muranga", "nairobi"],
    "salary": [5000, 6000, 2500, 4500]
}
df = pd.DataFrame(data)
print(df)
```
